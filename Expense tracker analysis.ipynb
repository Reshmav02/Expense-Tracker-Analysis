# Running SQL Queries

# 1. What is the total amount spent in each category?

```python
from tabulate import tabulate
query1= "SELECT Category, SUM(Amount) AS total_spent FROM expense_data GROUP BY Category;"
cursor_mysql.execute(query1)
res1=cursor_mysql.fetchall()

# Define table headers
headers=['Category', 'Total spent']

#printe the result as a table
print(tabulate(res1, headers=headers, tablefmt="grid"))
```

# 2. What is the total amount spent using each payment mode?

```python
from tabulate import tabulate
query2= "SELECT Payment_modes, SUM(amount) AS total_spent FROM expense_data GROUP BY Payment_modes;"
cursor_mysql.execute(query2)
res2=cursor_mysql.fetchall()

# Define table headers
headers=['Payment Mode', 'Total Spent amount']

#printe the result as a table
print(tabulate(res2, headers=headers, tablefmt="grid"))
```

# 3. Which are the top 5 most expensive categories in terms of spending?

```python
from tabulate import tabulate
query3= "SELECT Category, SUM(Amount) AS total_spent FROM expense_data GROUP BY Category ORDER BY total_spent DESC LIMIT 5;"
cursor_mysql.execute(query3)
res3=cursor_mysql.fetchall()

# Define table headers
headers=['Category', 'Total Spent amount']

#printe the result as a table
print(tabulate(res3, headers=headers, tablefmt="grid"))
```

# 4. How much was spent on subscripton using different payment modes?

```python
query4= "SELECT Payment_modes, SUM(amount) AS total_spent FROM expense_data WHERE Category = 'subscription' GROUP BY Payment_modes;"
cursor_mysql.execute(query4)
res4=cursor_mysql.fetchall()

# Define table headers
headers=['List', 'Total Spent amount']

#printe the result as a table
print(tabulate(res4, headers=headers, tablefmt="grid"))
```

# 5. How much was spent on stationery using different payment modes?

```python
query5= "SELECT Payment_modes, SUM(amount) AS total_spent FROM expense_data WHERE Category = 'stationery' GROUP BY Payment_modes;"
cursor_mysql.execute(query5)
res5=cursor_mysql.fetchall()

# Define table headers
headers=['Payment mode List', 'Total Spent amount']

#printe the result as a table
print(tabulate(res5, headers=headers, tablefmt="grid"))
```

# 6. How much amount was spent using UPI mode based on categories?

```python
query6= "SELECT  Category,  SUM(Amount) AS total_upi_spent FROM  expense_data WHERE  Payment_modes = 'UPI' GROUP BY  Category ORDER BY total_upi_spent DESC;"
cursor_mysql.execute(query6)
res6=cursor_mysql.fetchall()

# Define table headers
headers=['Category', 'Total Spent amount in UPI']

#printe the result as a table
print(tabulate(res6, headers=headers, tablefmt="grid"))
```

# Creating a Streamlit App with Filters

```python
%%writefile app.py
import streamlit as st
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to connect to SQLite database
def get_data(query, params=None):
    conn = sqlite3.connect("expense_database.sqlite")
    if params:
        df = pd.read_sql_query(query, conn, params=params)
    else:
        df = pd.read_sql_query(query, conn)
    conn.close()
    return df

# Streamlit App Title
st.set_page_config(page_title="Analyzing Personal Expenses", layout="wide")

# Sidebar for navigation
st.sidebar.title("Navigation")
page = st.sidebar.radio("Go to", ["Project Introduction", "Expense Data Visualization", "SQL Queries", "Creator Info"])

# -------------------------------- PAGE 1: Introduction --------------------------------
if page == "Project Introduction":
    st.title("Analyzing Personal Expenses")
    st.image("expense_tracker.jpg")    #Adding image
    st.subheader("📊 A Streamlit App for Exploring Personal Expenses")
    st.write("""
    The Analyzing Personal Expenses project focuses on tracking and categorizing individual spending patterns over time. 
    It aims to provide insights into monthly expenses, highlight spending trends, and identify areas for potential savings. 
    The project leverages data visualization and statistical analysis to support informed financial decisions.
    
    
    **Features:**
    - Generates monthly summaries showing total spending and averages by category.
    - Provides charts and graphs to visualize spending patterns over time.
    - Run predefined SQL queries to explore insights.
    
    **Database Used:** `expense_database.sqlite`
    """)

# -------------------------------- PAGE 2: Weather Data Visualization --------------------------------
elif page == "Expense Data Visualization":
    st.title("📊 Expense Data Visualizer")
    
    # Fetch categories list
    categories = get_data("SELECT DISTINCT Category FROM expense_data")["Category"].tolist()
    
    # Filters
    selected_category = st.selectbox("Select Category", categories)
    date_option = st.radio("Filter By:", ["Specific Day", "Entire Month"])
    
    if date_option == "Specific Day":
        selected_date = st.date_input("Choose a Date")
        query = "SELECT * FROM expense_data WHERE Category = ? AND Date = ?"
        df = get_data(query, params=(selected_category, selected_date.strftime("%Y-%m-%d")))
    else:
        selected_month = st.selectbox("Select Month", range(1, 13))
        query = "SELECT * FROM expense_data WHERE Category = ? AND strftime('%m', Date) = ?"
        df = get_data(query, params=(selected_category, f"{selected_month:02d}"))

    if not df.empty:
        st.write("### Expense Data", df)
        
        # Select feature to visualize
        feature = st.selectbox("Select Feature to Visualize", 
            ["Category", "Payment_modes"])

        # Visualization
        st.write(f"### {feature} Trends")
        plt.figure(figsize=(10, 5))
        sns.lineplot(data=df, x="Date", y=feature, marker="o", label=feature)
        plt.xticks(rotation=45)
        st.pyplot(plt)

        # Bar chart for Weather Condition
        st.write("### Category Distribution")
        plt.figure(figsize=(8, 4))
        sns.countplot(y=df["Category"], palette="coolwarm")
        st.pyplot(plt)

    else:
        st.warning("No data available for the selected filters.")

# -------------------------------- PAGE 3: SQL Queries --------------------------------
elif page == "SQL Queries":
    st.title("📋 SQL Query Results")
    
    queries = {
        "1. Total amount spent in each category?": "SELECT Category, SUM(Amount) AS total_spent FROM expense_data GROUP BY Category",
        "2. Total amount spent using each payment mode?": "SELECT Payment_modes, SUM(amount) AS total_spent FROM expense_data GROUP BY Payment_modes",
        "3. Top 5 most expensive categories in terms of spending?": "SELECT Category, SUM(Amount) AS total_spent FROM expense_data GROUP BY Category ORDER BY total_spent DESC LIMIT 5",
        "4. Amonut spent on subscripton using different payment modes": "SELECT Payment_modes, SUM(amount) AS total_spent FROM expense_data WHERE Category = 'subscription' GROUP BY Payment_modes",
        "5. Amount spent on stationery using different payment modes": "SELECT Payment_modes, SUM(amount) AS total_spent FROM expense_data WHERE Category = 'stationery' GROUP BY Payment_modes"
        "6. Amount was spent using UPI mode based on categories": "SELECT  Category,  SUM(Amount) AS total_upi_spent FROM  expense_data WHERE  Payment_modes = 'UPI' GROUP BY  Category ORDER BY total_upi_spent DESC"
    }
    
    selected_query = st.selectbox("Choose a Query", list(queries.keys()))
    query_result = get_data(queries[selected_query])
    
    st.write("### Query Result:")
    st.dataframe(query_result)

# -------------------------------- PAGE 4: Creator Info --------------------------------
elif page == "Creator Info":
    st.title("👩‍💻 Creator of this Project")
    st.write("""
    **Developed by:** Reshma V 
    **Skills:** Python, SQL, Data Analysis,Streamlit, Pandas   
    """)


```
